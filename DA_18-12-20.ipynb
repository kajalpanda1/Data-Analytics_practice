```
import numpy as np
arr1=np.arange(10,100,10)
arr1=arr1.reshape((3,3))
arr2=np.arange(1,10)
arr2=arr2.reshape((3,3))
print('Array1 is \n',arr1)
print('Array2 is \n',arr2)
```

```
add=np.add(arr1,arr2)
sub=np.subtract(arr1,arr2)
mul=np.multiply(arr1,arr2)
div=np.divide(arr1,arr2)
print('Sum of array is \n',add)
print('Sub of array is \n',sub)
print('Mul of array is \n',mul)
print('Div of array is \n',div)
```
