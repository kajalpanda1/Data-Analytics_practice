```
moviedata=pd.read_csv('Movie_Id_Titles')
moviedata
```

```
colnames=['userid','item_id','ratings','time']
userdata=pd.read_csv('u.data',sep='\t',names=colnames)
userdata
```

```
df_movie=pd.merge(moviedata,userdata,how='inner',on='item_id')
df_movie
```

```
df_moviepivot=df_movie.pivot_table(index='userid',columns='title',values='ratings')
df_moviepivot
```

```
userid=int(input('enter user id'))
df_moviepivot.iloc[userid].dropna()
```

```
moviename=input('enter movie name')
df_moviepivot[moviename]
```

```
moviename=input('enter movie name').strip()
df_similarmovie=df_moviepivot.corrwith(df_moviepivot[moviename])
df_similar=pd.DataFrame(df_similarmovie,columns=['Correlated Ratings'])
df_similar
```

```
df_similar.sort_values(by=['Correlated Ratings'],ascending=False)
```

```
df_nmovie=pd.DataFrame(df_movie.groupby('title')['ratings'].count()[1:])
df_nmovie['Avg Ratings']=df_movie.groupby('title')['ratings'].mean()
#df_nmovie.sort_values(by='ratings',ascending=False)
df_nmovie.rename(columns={'ratings' :'number of ratings'})
```
