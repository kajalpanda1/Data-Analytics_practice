```
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```
df=pd.DataFrame(data=[134,49,123],index=['BMW','Audi','Mercedes'],columns=['Number of units sold'])
df
```

```
plt.bar(df.index,df['Number of units sold'])
plt.title('Car Sales')
plt.show()
```

```
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.bar(df.index,df['Number of units sold'],color=['r','b','g'])
plt.show()
```

```
df['Percent']=df['Number of units sold']/sum(df['Number of units sold']) * 100
df
```

```
df.sort_values('Number of units sold', ascending=False,inplace=True)
df
```

```
df['cumulative percent']=(df['Number of units sold'].cumsum()/df['Number of units sold'].sum())*100
df
```

```
from matplotlib.ticker import PercentFormatter
fig,axes=plt.subplots()
axes.bar(df.index,df['Number of units sold'],color='c')
axes2=axes.twinx()
axes2.plot(df.index,df['cumulative percent'],color='k',marker='s',ms=10)
axes2.yaxis.set_major_formatter(PercentFormatter())
axes.tick_params(axis='y',color='g')
axes2.tick_params(axis='y',color='r')
plt.show()
```

```
states=['Assam','Bihar','Chandigarh','Delhi','Goa']
cases=[248028,248017,19184,618747,50239]
plt.figure(figsize=(20,5))
plt.barh(states,cases)
plt.xlabel('Number of Covid Cases')
plt.ylabel('City Names')
plt.title('State wise-covid cases')
plt.xticks(cases)
plt.show()
```

```
df_car=pd.DataFrame([[30,25,50,20],[40,23,51,17],[35,22,45,19]],index=['BMW','Audi','Mercedes'],columns=['2001','2005','2010','2020'])
df_car
```

```
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
x=np.arange(4)
ax.bar(x+0.00,np.array(df_car.iloc[0:1,:]).ravel(),color='b', width=0.25)
ax.bar(x+0.25,np.array(df_car.iloc[1:2,:]).ravel(),color='r', width=0.25)
ax.bar(x+0.50,np.array(df_car.iloc[2:,:]).ravel(),color='y', width=0.25)
ax.set_title('Car Sales ')
ax.set_xlabel('Year')
ax.set_ylabel('No Of Sales')
ax.legend(labels=['BMW','Audi','Mercedes'])
plt.show()
```

```
x=np.arange(5)
oilration=(10,20,40,25,60)
gasration=(90,80,60,75,40)
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.set_title('Oil-Gas Ratio')
axes.set_ylabel('Units in ml')
axes.set_yticks(np.arange(0,101,10))

axes.bar(x,oilration,width=0.35,color='b')
axes.bar(x,gasration,width=0.35,color='y',bottom=oilration)
axes.legend(labels=['Oil','Gas'])
plt.show()
```

```
jobs=[29000,18000,24000,19000]
pl=['Java','C#','JS','Python']
df_data=pd.DataFrame(jobs,columns=['Number of jobs'],index=pl)
```

```
df_data['Percent']=df_data['Number of jobs']/sum(df_data['Number of jobs']) * 100
```

```
df_data
```

```
df_data.sort_values('Number of jobs', ascending=False,inplace=True)
```

```
plt.pie(df_data['Number of jobs'],labels=df_data.index,autopct='%.2f%%',shadow=True)
```

```
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.axis('equal')
axes.pie(df_data['Number of jobs'],labels=df_data.index,autopct='%.2f%%',shadow=True)
plt.show()
```

```
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.axis('equal')
axes.pie(df_data['Number of jobs'],labels=df_data.index,autopct='%.2f%%',shadow=True,explode=[0,0,0,0.2],colors=['blue','grey','#ff8899','#88ff99'])
plt.show()
```

```
df_data['cumulative job percent']=(df_data['Number of jobs'].cumsum()/df_data['Number of jobs'].sum())*100
df_data
```

```
from matplotlib.ticker import PercentFormatter
fig,axes=plt.subplots()
axes.bar(df_data.index,df_data['Number of jobs'],color='#8899ff')
axes2=axes.twinx()
axes2.plot(df_data.index,df_data['cumulative job percent'],color='r',marker='s',ms=7)
axes2.yaxis.set_major_formatter(PercentFormatter())
axes.tick_params(axis='y',color='g')
axes2.tick_params(axis='y',color='r')
plt.show()
```
