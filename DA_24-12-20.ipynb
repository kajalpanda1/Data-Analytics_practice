```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df_data=pd.DataFrame(data=[10,20,30,40,50],columns=['Price'])
```

```
df_data
```

```
mean=np.mean(df_data['Price'])
```

```
df_data['Mean Sq Difference']=(df_data.Price-mean)**2
```

```
df_data
```

```
var=df_data['Mean Sq Difference'].sum()/len(df_data['Price'])-1
```

```
var
```

```
plt.bar(df_data.index,df_data['Mean Sq Difference'])
plt.show()
```

```
import math
```

```
math.sqrt(var)
```

```
df_curr=pd.DataFrame(np.arange(1,11),columns=['Price in $'])
df_curr['Price in INR']=df_curr['Price in $'] * 73.55
```

```
df_curr
```

```
dmean=np.mean(df_curr['Price in $'])
dmean
```

```
rmean=np.mean(df_curr['Price in INR'])
rmean
```

```
df_curr['Mean Diff $']=df_curr['Price in $']-dmean
```

```
df_curr['Mean Diff INR']=df_curr['Price in INR']-rmean
```

```
df_curr
```

```
covar=(df_curr['Mean Diff $']*df_curr['Price in INR']).sum()/len(df_curr)-1
```

```
covar
```

```
dstd=np.std(df_curr['Price in $'])
print('Std of dollar ',dstd)
```

```
rstd=np.std(df_curr['Price in INR'])
print('Std of rupee ',rstd)
```

```
cvrcoff=covar/(rstd*dstd)
cvrcoff
```
