##Create a matrix of 9x9. divide it into nine 3x3 matrics. take maximum out of each matrix and make one 3x3 matrix.

```
import numpy as np
import pandas as pd

matrix = np.random.randint(100, size=(9,9))

print("Matrix of shape (9,9):\n",matrix)

mat1=matrix[:3,0:3]
mat2=matrix[:3,3:6]
mat3=matrix[:3,6:9]

print(mat1,'\n', mat2,'\n', mat3)

mat4=matrix[3:6,0:3]
mat5=matrix[3:6,3:6]
mat6=matrix[3:6,6:9]

print(mat4,'\n', mat5,'\n', mat6)

mat7=matrix[6:,0:3]
mat8=matrix[6:,3:6]
mat9=matrix[6:,6:9]

print(mat7,'\n', mat8,'\n', mat9)

new_mat=np.array([np.max(mat1), np.max(mat2), np.max(mat3), np.max(mat4), np.max(mat5), np.max(mat5), np.max(mat6), np.max(mat7), np.max(mat8)])

print("New matrix of shape(3,3):\n",new_mat.reshape(3,3))
```
